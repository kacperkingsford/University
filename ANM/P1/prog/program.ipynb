```julia
function suma_rosnaca(n) #funkcja obliczająca sume w porządku naturalnym
    suma = 0.0
    for k in 0:n
        if k%2==0
            suma+=1/(2*k + 1)
        else
            suma-=1/(2*k + 1)
        end
    end
    4.0*suma
end

function suma_malejaca(n) #funkcja obliczająca sume w porządku odwrotnym
    suma = 0.0
    for k in reverse(0:n)
        if k%2==0
            suma+=1/(2*k + 1)
        else
            suma-=1/(2*k + 1)
        end
    end
    4.0*suma
end

function relativeError(f,n) #funkcja obliczająca błąd względny
    pi = f(Float32(n))
    Δpi = Float32(π) - pi
    Float32(abs(Δpi/pi))
end


function test(f, n) #funkcja obliczająca ilość cyfr znaczących, błąd względny obu funkcji
    pi = f(Float32(n))
    relative_error = relativeError(f,n)
    accuracy = floor(-log10(relative_error))
    println("n=$n:\n$pi\nSignif. digits: $accuracy\n relativeerror: $relative_error\n")
end

function relativeErrorRatio(n) #funkcja obliczająca stosunek błędów względnych dwóch funkcji
    ratio = relativeError(suma_rosnaca,n) / relativeError(suma_malejaca,n)
    println("n=$n, ratio=$ratio \n")
end

#funkcje testujące

for k in 1:7
    test(suma_rosnaca, 10^k)
end

for k in 1:7
    test(suma_malejaca, 10^k)
end

println("======================================================")

for k in 1:7
    relativeErrorRatio(10^k)
end

println("======================================================")
```

```julia
using Plots #pakiet potrzebny do rysowania wykresów
plotly()
```

```julia
# Wykres wyników funkcji suma_rosnaca(n) dla n = 10^1, 10^2, ..., 10^7
xs = 1:7
range = map(x -> BigFloat(10.0)^x, xs)
approx = map(x -> suma_rosnaca(x), range)
p = plot(
    xs, 
    xticks=1:1:7, 
    yticks=3.1:0.05:3.5, 
    ylims=(3.1, 3.5),
    map(x -> Float32(MathConstants.pi), xs),
    seriestype="line",
    label = "Dokładna wartość pi"
)
scatter!(
    xs,
    approx,
    label = "Przybliżona wartość pi"
)
```

```julia
function CalculaterelativeErrorRatio(n) #funkcja obliczająca iloraz błędów względnych obu funkcji
    ratio = relativeError(suma_rosnaca,n) / relativeError(suma_malejaca,n)
    ratio
end
```

```julia
#wykres funkcji suma_rosnaca i suma_malejaca, zauważmy, że wykresy są tak podobne, że nie widać między nimi różnicy
plot(
    0:100, 
    xticks=0:10:100, 
    yticks=3.1:0.1:3.5, 
    xlims=(0, 100),
    ylims=(3.1, 3.5),
    [
        map(x -> Float32(MathConstants.pi), 0:100),
        map(x -> suma_rosnaca(Float32(2*x)), 0:100),
        map(x-> suma_malejaca(Float32(2*x)), 0:100)
    ],
    xlabel="\$n\$", 
    ylabel="\$\\widetilde{pi}\$", 
    label = ["Dokładna wartość pi" "suma_rosnaca(n)" "suma_malejaca(n)"],
    lw = 1
)
```
